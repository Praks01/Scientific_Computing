#### Explicit Differentiation


```python
f(x)=erf(x)
df(x)=f.derivative(x)
df.show()
```

```python
fig1a=plot(f,(-3,3),frame=True, color='red',legend_label='erf(x)')
fig1b=plot(df,(-3,3),color='blue',legend_label='d erf(x)')
(fig1a+fig1b).show()
```

```python
g(x)=x^3-x+2
gprime(x)=g.derivative(x)
fig2a=plot(g,(-3,3),frame=True, color='red',legend_label='g')
fig2b=plot(gprime,(-3,3),color='blue',legend_label='gprime')
(fig2a+fig2b).show()
```

```python
h(x)=function('h')(x)
eqn=5*h^2+sin(h)==x^2
eqn.show()
```

```python
dhdx=solve(diff(eqn), diff(h))
dhdx
```

### Taylor Series

```python
p(x)=sin(x)/x^2
p.show()
```

```python
plot(p,(x,1,12))
```

```python
pt(x)=p.taylor(x,3,5)
pt.show()
```

```python
fig1=plot(p,(x,1,5),color='red',legend_label='p')
fig2=plot(pt,(x,1,5),color='green',legend_label='pt')
(fig1+fig2).show()
```

### Fourier series

```python
f=piecewise([((0,pi/2),-1),((pi/2,3*pi/4),0),((3*pi/4,pi),2)])
```

```python
fig4a=plot(f,(0,pi))
fig4a.show()
```

```python
f.fourier_series_cosine_coefficient(1)
f.fourier_series_sine_coefficient(10)
ff=f.fourier_series_partial_sum(20)
ff.show()
```

```python
fig4b=plot(ff,(0,pi),color='red')
(fig4a+fig4b).show()
```
